```python
import pandas as pd
import numpy as np
import folium 
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)[0]
df.columns = ['Postal Code', 'Borough', 'Neighourhood']
df = df[df["Borough"] != "Not assigned"]
df = df.groupby('Postal Code').agg({'Borough' : 'first', 'Neighourhood' : ','.join}).reset_index().reindex(columns=df.columns)
df.loc[df['Neighourhood'] == 'Not assigned', 'Neighourhood'] = df['Borough']
```

```python
geo_df = pd.read_csv("https://cocl.us/Geospatial_data")
df = pd.merge(df, geo_df, on='Postal Code')
df.head()
```

## First Visualisation
First we visualise all neighbourhoods

```python

cmap = folium.Map(location=[43.651070, -79.347015], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(cmap)  
cmap
```

## Boroughs that contain the word "Toronto"
Now to simplify the view slightly, we only visualise neighbourhoods that contain the word Toronto in the borough

```python
toronto_df = df[df['Borough'].str.contains("Toronto")]
toronto_df.head()
```

```python
toronto_cmap = folium.Map(location=[43.651070, -79.347015], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(toronto_df['Latitude'], toronto_df['Longitude'], toronto_df['Borough'], toronto_df['Neighourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_cmap)  
toronto_cmap
```
